```python
import pandas as pd

train_csv_path = "./data/train/train_agent"
#train_csv = pd.read_csv(train_csv_path)

# list of tuples
# (token, count)
arg_hist = {
    '0': {},
    '1': {},
    '2': {},
    '3': {},
    '4': {}
}

#k = [1,2,3,4,5]
:q

#for i in range(len(train_csv)):
    #with open("./data/train/" + train_csv.iloc[i,0] + "." + train_csv.iloc[i,1]) as e:
with open(train_csv_path) as e:

    #read all lines and then process each
    lines = e.readlines()

    for l in lines:

        if not '#' in l: # ignore comments

            all_args = l.split('|')[1].strip() # all args are after | 
            arg_list = all_args.split(' ') # args are space separated

            # go over all k args and increment the counts
            for k in range(len(arg_list)):
                acc = arg_hist[str(k)][arg_list[k]] if arg_list[k] in arg_hist[str(k)] else 0
                arg_hist[str(k)][arg_list[k]] = acc + 1 
                if k == 4: break


            
            
            
```
